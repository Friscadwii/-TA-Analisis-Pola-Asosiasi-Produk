# MARKET BASKET ANALISIS (MBA)

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Menghitung Jumlah Produk lebih dari satu produk
grup = df_grouped.groupby("No. Pesanan")["Kode Produk"].nunique()
pesanan = grup[grup > 1].index
filter_produk = df_grouped[df_grouped["No. Pesanan"]. isin(pesanan)]
filter_produk
```

```python
# Simpan ke Excel
filter_produk.to_excel("Produk lebih 1.xlsx", index=False)
```

```python
#Membentuk pivot table untuk basket analysis
basket = (
    filter_produk.groupby(['No. Pesanan', 'Kode Produk'])['Jumlah']
    .sum()
    .reset_index()
    .pivot(index='No. Pesanan', columns=['Kode Produk'], values='Jumlah')
    .fillna(0)
)
basket
```

```python
#One Hot-Encode
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
basket
```

```python
# Simpan ke Excel
basket.to_excel("Endcoding.xlsx", index=False)
```

```python
from itertools import combinations
import pandas as pd
produk_list = basket.columns.tolist()
kombinasi_produk = list(combinations(produk_list, 2))
kombinasi_produk
```

```python
kombinasi_count = []
for item1, item2 in kombinasi_produk:
    # Hitung transaksi di mana kedua item dibeli bersama
    count = ((basket[item1] == 1) & (basket[item2] == 1)).sum()
    kombinasi_count.append({
        'Item 1': item1,
        'Item 2': item2,
        'Jumlah Muncul Bersama': count
    })

# 4. Ubah menjadi DataFrame
kombinasi_df = pd.DataFrame(kombinasi_count)
kombinasi_df
```

```python
kombinasi_df.to_excel("kombinasi_2item.xlsx", index=False)
```

```python
frequent_itemsets = apriori(basket, min_support=0.03, use_colnames=True)
frequent_itemsets
```

```python
# Simpan ke Excel
frequent_itemsets.to_excel("Frekuensi.xlsx", index=False)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# lift 0.5
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

# Hanya (P002, P006)
filtered_rules = rules[
    rules['antecedents'].apply(lambda x: {'P002'}.issubset(x)) &
    rules['consequents'].apply(lambda x: {'P006'}.issubset(x))
]
filtered_rules[["support", "confidence", "lift"]]
```

```python
result = rules[(rules['confidence'] >= 0.3) &
               (rules['lift'] >= 2.0)]
result
```

```python
frequent_itemsets3 = apriori(basket, min_support=0.001, use_colnames=True, max_len=3)
# 3 pola
rules3 = association_rules(frequent_itemsets3, metric="lift", min_threshold=1)
rules3
```
